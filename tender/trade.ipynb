```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from CSV file into a DataFrame
df = pd.read_csv('procurement-notices.csv')

# Set the Seaborn style
sns.set_style('darkgrid')

# Create a bar plot of the Notice Type counts
sns.countplot(x='Notice Type', data=df)
plt.title('Notice Type Counts')
plt.xlabel('Notice Type')
plt.ylabel('Count')
plt.show()

# Create a scatter plot of the Publication Date vs. Deadline Date
sns.scatterplot(x='Publication Date', y='Deadline Date', data=df)
plt.title('Publication Date vs. Deadline Date')
plt.show()

# Create a box plot of the Procurement Type by Country Name
sns.boxplot(x='Procurement Type', y='Country Name', data=df)
plt.title('Procurement Type by Country Name')
plt.xlabel('Procurement Type')
plt.ylabel('Country Name')
plt.show()
```
